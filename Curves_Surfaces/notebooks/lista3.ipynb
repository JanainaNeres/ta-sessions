**Exercício 1.** Verificar regularidade, comprimento de arco e curvatura.

- Reta

```python
a, b, c, d, t = var('a'), var('b'), var('c'), var('d'), var('t')
assume(t,'real')
alpha = vector((a + c*t, b + d*t))

# a = 1, b = 2, c = 2, d = 3
parametric_plot((1 + 2*t, 2 + 3*t), (t,-2,2), rgbcolor=hue(0.6))
```

```python
# é regular? 
gradient = diff(alpha, t)
gradient
```

```python
# Comprimento de arco
def integrand(u):
    alpha = vector((a + c*u, b + d*u))
    return norm(diff(alpha, u))

T = var('T')

definite_integral(integrand(t), t, 0, T)
```

```python
# Defining operator J
def J(v):
    return vector((-v[1], v[0]))
```

```python
# Curvatura
(J(gradient)).dot_product(diff(gradient, t))
```

- $\alpha(t) = (t, t^4)$

```python
alpha = vector((t, t^4))
parametric_plot(alpha, (t,-1,1), rgbcolor=hue(0.6))
```

```python
# é regular? 
gradient = diff(alpha, t)
gradient
```

```python
# Curvatura
(J(gradient)).dot_product(diff(gradient, t))/norm(gradient)^3
```

- $\alpha(t) = \cos(t)(2 \cos(t) − 1), \sin(t)(2 \cos(t) − 1))$

```python
alpha = (2*cos(t) - 1) * vector((cos(t), sin(t)))

figures = []
figures.append(parametric_plot(alpha, (t,0,2*pi), rgbcolor=hue(0.6)))
figures.append(plot((diff(alpha, t)).subs(t=1), color='red'))
sum(figures)
```

```python
# é regular? 
gradient = diff(alpha, t)
gradient
```

```python
norm(diff(alpha, t))
```

```python
# Comprimento de arco
def integrand(u):
    alpha = (2*cos(u) - 1) * vector((cos(u), sin(u)))
    return (norm(diff(alpha, u))).simplify_full()

definite_integral(integrand(t), t, 0, 2*pi)
```
